# Regular Expression Concepts in Python DataFrame (Pandas)

### We mainly use regex in Pandas with:

### 1. Basic Pattern Matching
##### `.` → Any character

```python
import pandas as pd

df = pd.DataFrame({
    'Name': ['Ram', 'Rim', 'Rom', 'Rkm']
})

df[df['Name'].str.contains('R.m')]

# Matches: Ram, Rim, Rom
```

## 2. Starts With `^`

```python
df = pd.DataFrame({'City': ['Delhi', 'Mumbai', 'Dehradun', 'Pune']})

df[df['City'].str.contains('^De')]

# Matches: Delhi, Dehradun
```

## 3. Ends With `$`

```python
df = pd.DataFrame({'City': ['Delhi', 'Mumbai', 'Dehradun', 'Pune']})

df[df['City'].str.contains('ne$')]

# Matches: Pune
```

## 4. Digit `\d`

```python
df = pd.DataFrame({'Code': ['A123', 'B456', 'C789', 'ABC']})

df[df['Code'].str.contains('\d')]

#Matches rows containing numbers.
```

## 5. Non-Digit `\D`

```python
df = pd.DataFrame({'Code': ['A123', 'B456', '789', 'ABC']})

df[df['Code'].str.contains('\D')]
```

## 6. Alphabets `[a-z]`

```python
df = pd.DataFrame({'Code': ['A123', 'B456', '789', 'ABC']})

df[df['Code'].str.contains('[A-Z]')]
```

## 7. Multiple Characters `[abc]`

```python
df = pd.DataFrame({
    'Name': ['Ram', 'Rim', 'Rom', 'Rkm']
})

df[df['Name'].str.contains('[aeiou]')]

#Match a OR b OR c
```

## 8. Not in List `[^abc]`

```python
df = pd.DataFrame({
    'Name': ['Ram', 'Rim', 'Rom', 'Rkm']
})

df[df['Name'].str.contains('[^aeiou]')]
```

## 9. OR Operator `|`

```python
df = pd.DataFrame({'City': ['Delhi', 'Mumbai', 'Dehradun', 'Pune']})

df[df['City'].str.contains('Delhi|Pune')]
```

## 10. Zero or More `*`

```python
df = pd.DataFrame({'Text': ['ab', 'aab', 'aaab', 'b', '21']})

df[df['Text'].str.contains('a*b')]
```

## 11. One or More `+`

```python
df[df['Text'].str.contains('a+b')]
```

## 12. Optional `?`

```python
df[df['Text'].str.contains('a?b')]
```

## 13. Exact Number `{n}`

```python
df = pd.DataFrame({'Num': ['123', '1234', '12']})

df[df['Num'].str.contains('\d{3}')]
```

## 14. Range `{n,m}`

```python
df[df['Num'].str.contains('\d{2,3}')]
```

# 15. Word Boundary `\b`

```python
df = pd.DataFrame({'Text': ['cat', 'scatter', 'cat dog']})

df[df['Text'].str.contains(r'\bcat\b')]

# Matches exact word "cat".
```

## 16. Extract Data using Regex

```python
df = pd.DataFrame({'Email': ['abc@gmail.com', 'xyz@yahoo.com']})

df['Username'] = df['Email'].str.extract(r'(\w+)@')
df
```

## 17. Replace using Regex

```python
df = pd.DataFrame({'Email': ['abc@gmail.com', 'xyz@yahoo.com']})

df['Email'] = df['Email'].str.replace(r'@.*', '', regex=True)
df
```

## 18. Find All Matches

```python
df = pd.DataFrame({'Text': ['abc123', 'def456']})
df['Numbers'] = df['Text'].str.findall('\d+')
df
```

## 19. Match Pattern from Start

```python
df = pd.DataFrame({'Text': ['abc123', 'def456', "123adc"]})

df[df['Text'].str.match('[a-z]+')]
```
