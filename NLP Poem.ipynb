## Raw Text → Cleaning → Tokenization → Stopword Removal → Stemming/Lemmatization → POS/NER → Vectorization (BoW / TF-IDF) → Model

## Step-by-step Explanation
1. ### Raw Text
    - Original unprocessed text data.

2. ### Text Cleaning
    - Lowercasing
    - Removing punctuation, numbers, special characters

3. ### Tokenization
    - Splitting text into words (tokens)

4. ### Stopword Removal
    - Removing common words like: `is, am, are, the, and, etc.`

5. ### Stemming / Lemmatization
    - Stemming: Cutting words `(e.g., playing → play)`
    - Lemmatization: Dictionary-based `(e.g., better → good)`

6. ### POS Tagging / NER
    - POS Tagging (Part of Speech) → Grammar tagging `(noun, verb, adjective, etc.)` 
    - NER (Named Entity Recognition) → Detect names, places, organizations

7. ### Vectorization
    - Convert text → numbers:
        - One-Hot Encoding
        - Bag of Words (BoW)
        - TF-IDF (Term Frequency–Inverse Document Frequency)

8. ### ML / DL Model
    - Logistic Regression
    - Naive Bayes
    - LSTM / BERT (Deep Learning)

```python
# Raw Text
Text ='''
The Zen of Python, by Tim Peters

Beautiful is better than ugly.
Explicit is better than implicit.
Simple is better than complex.
Complex is better than complicated.
Flat is better than nested.
Sparse is better than dense.
Readability counts.
Special cases aren't special enough to break the rules.
Although practicality beats purity.
Errors should never pass silently.
Unless explicitly silenced.
In the face of ambiguity, refuse the temptation to guess.
There should be one-- and preferably only one --obvious way to do it.
Although that way may not be obvious at first unless you're Dutch.
Now is better than never.
Although never is often better than *right* now.
If the implementation is hard to explain, it's a bad idea.
If the implementation is easy to explain, it may be a good idea.
Namespaces are one honking great idea -- let's do more of those!
'''
print(Text)
```

```python
# Text Cleaning
import re # Liberary part
clean_text=Text.lower()
clean_text=re.sub(r'[^a-z\s]',"",clean_text)
print(clean_text)
```

```python
# Tokkenwise word
from nltk.tokenize import word_tokenize
tokens=word_tokenize(clean_text)
print(tokens)
```

```python
# Tokkenwise sentences
from nltk.tokenize import sent_tokenize
sentences=sent_tokenize(clean_text)
print(sentences)
```

```python
# stopwords
from nltk.corpus import stopwords
stop_word=set(stopwords.words('english'))
print(stop_word)
```

```python
# Stopword Remove
filterd_tokens = [word for word in tokens if word not in stop_word]
print(filterd_tokens)
```

```python
# Steming 
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed_word=[stemmer.stem(word) for word in filterd_tokens]
print(stemmed_word)
```

```python
 # Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemma_words = [lemmatizer.lemmatize(word) for word in filterd_tokens]
print(lemma_words)
```

```python
# POS Tagging (Part of Speech) 
tags = nltk.pos_tag(lemma_words)
print(tags)
```

```python
# NER (Named Entity Recognition)
name_entities = nltk.ne_chunk(nltk.pos_tag(tokens))
print(name_entities)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# 1. Ensure all items are strings (handles tokens, tuples, or trees)
cleaned_words = []
for item in tokens:
    if isinstance(item, str):
        cleaned_words.append(item)
    elif isinstance(item, tuple):
        cleaned_words.append(item[0])
    else:
        # Handles NLTK Tree objects if they exist
        cleaned_words.extend([leaf[0] for leaf in item.leaves()])

# 2. Join the words into a single "document" string
# CountVectorizer expects a list of documents
text_data = [" ".join(cleaned_words)]

# 3. Initialize and transform
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(text_data)

# 4. View results
print("Feature Names:", vectorizer.get_feature_names_out())
print("Array:\n", x.toarray())
```

```python
# TF-IDF (Term Frequency–Inverse Document Frequency)
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Clean the data (same logic as the CountVectorizer part)
# This ensures we are passing a list of strings, not tuples or objects
cleaned_txt = []
for item in tokens:
    if isinstance(item, str):
        cleaned_txt.append(item)
    elif isinstance(item, tuple):
        cleaned_txt.append(item[0])
    else:
        cleaned_txt.extend([leaf[0] for leaf in item.leaves()])

# 2. Join into a document and wrap in a list
formatted_txt = [" ".join(cleaned_txt)]

# 3. Initialize and transform
tfidf = TfidfVectorizer()
y = tfidf.fit_transform(formatted_txt)

# 4. View results
print("Feature Names:", tfidf.get_feature_names_out())
print("TF-IDF Matrix:\n", y.toarray())
```
