## Raw Text → Cleaning → Tokenization → Stopword Removal → Stemming/Lemmatization → POS/NER → Vectorization (BoW / TF-IDF) → Model

## Step-by-step Explanation
1. ### Raw Text
    - Original unprocessed text data.

2. ### Text Cleaning
    - Lowercasing
    - Removing punctuation, numbers, special characters

3. ### Tokenization
    - Splitting text into words (tokens)

4. ### Stopword Removal
    - Removing common words like: `is, am, are, the, and, etc.`

5. ### Stemming / Lemmatization
    - Stemming: Cutting words `(e.g., playing → play)`
    - Lemmatization: Dictionary-based `(e.g., better → good)`

6. ### POS Tagging / NER
    - POS Tagging (Part of Speech) → Grammar tagging `(noun, verb, adjective, etc.)` 
    - NER (Named Entity Recognition) → Detect names, places, organizations

7. ### Vectorization
    - Convert text → numbers:
        - One-Hot Encoding
        - Bag of Words (BoW)
        - TF-IDF (Term Frequency–Inverse Document Frequency)

8. ### ML / DL Model
    - Logistic Regression
    - Naive Bayes
    - LSTM / BERT (Deep Learning)

```python
# Raw Text
Text ='''
Natural language processing (NLP) is amazing!
It allows machine to understand natural Language.
I love Learning NLP.
'''
print(Text)
```

```python
# Text Cleaning
import re # Liberary part
clean_text=Text.lower()
clean_text=re.sub(r'[^a-z\s]',"",clean_text)
print(clean_text)
```

```python
# Tokkenwise word
from nltk.tokenize import word_tokenize
tokens=word_tokenize(clean_text)
print(tokens)
```

```python
# Tokkenwise sentences
from nltk.tokenize import sent_tokenize
sentences=sent_tokenize(clean_text)
print(sentences)
```

```python
# stopwords
from nltk.corpus import stopwords
stop_word=set(stopwords.words('english'))
print(stop_word)
```

```python
# Stopword Remove
filterd_tokens = [word for word in tokens if word not in stop_word]
print(filterd_tokens)
```

```python
# Steming 
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
stemmed_word=[stemmer.stem(word) for word in filterd_tokens]
print(stemmed_word)
```

```python
 # Lemmatization
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lemma_words = [lemmatizer.lemmatize(word) for word in filterd_tokens]
print(lemma_words)
```

```python
# POS Tagging (Part of Speech) 
tags = nltk.pos_tag(lemma_words)
print(tags)
```

```python
# NER (Named Entity Recognition)
name_entities = nltk.ne_chunk(nltk.pos_tag(tokens))
print(name_entities)
```

```python
txt = ['i love nlp', 'nlp is amazing nlp is good', 'i love learninng' ]
```

```python
# Vectorization
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
x = vectorizer.fit_transform(txt)
print(x)
```

```python
print(vectorizer.get_feature_names_out())
print(x.toarray())
```

```python
# TF-IDF (Term Frequency–Inverse Document Frequency)
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
y = tfidf.fit_transform(txt)
print(y)
```

```python
print(tfidf.get_feature_names_out())
print(y.toarray())
```
